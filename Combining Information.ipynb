# Combining information

```python
import numpy as np
import pandas as pd
from sklearn import datasets
```

```python
titanic = datasets.fetch_openml(name='titanic', version=1)
df = pd.DataFrame(titanic.data, columns = titanic.feature_names)
df['target'] = titanic.target
```

```python
df
```

```python
df.drop(columns = ['pclass', 'name', 'fare', 'ticket', 'boat', 'embarked', 'body', 'target', 'home.dest'], axis=1, inplace=True)
```

```python
df['alone'] = np.where((df['sibsp']+df['parch'])>0,1,0)
```
